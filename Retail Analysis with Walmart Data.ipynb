##### Provide a monthly and semester view of sales in units and give insights

```python
Walmart['year']= Walmart['Date'].dt.year
Walmart['month']= Walmart['Date'].dt.month
Walmart['month_year']= list(zip(Walmart['month'],Walmart['year']))
Walmart['month_year']
```

```python
def semester(num):
     if num in [(2,2010),(3,2010),(4,2010),(5,2010),(6,2010)]:
        return 1
     elif num in [(7,2010),(8,2010),(9,2010),(10,2010),(11,2010),(12,2010)]:
        return 2
     elif num in [(2,2011),(3,2011),(4,2011),(5,2011),(6,2011)]:
        return 3
     elif num in [(7,2011),(8,2011),(9,2011),(10,2011),(11,2011),(12,2011)]:
        return 4
     elif num in [(2,2012),(3,2012),(4,2012),(5,2012),(6,2012)]:
        return 5
     else:
        return 6
```

```python
Walmart['semester'] = Walmart['year_month'].apply(lambda x: semester(x))
sns.lineplot(x = 'month', y = 'Weekly_Sales',
             hue = 'year',
            data = Walmart,
            ci = None,
            estimator = np.sum,
             palette = 'pastel')
```

```python
sns.lineplot(x = 'semester', y = 'Weekly_Sales',
            data = Walmart,
            ci = None,
            estimator = np.sum)
```

#### Statistical Model



```python
x_features_object= Walmart[Walmart['Store']==1][['Store','Date']]
date_obj= Walmart[Walmart['Store']==1][['Date']]
date_obj.index+=1
x_features_object.Date=date_obj.index
x_features_object.head()
```

```python
y_target= Walmart[Walmart['Store']==1]['Weekly_Sales']
y_target.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_features_object, y_target, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
linreg= LinearRegression()
linreg.fit(x_train, y_train)
feature_dataset= Walmart[Walmart['Store']==1][['Store','CPI','Unemployment','Fuel_Price']]
feature_dataset.head()
```

```python
response_cpi= Walmart[Walmart['Store']==1]['CPI'].astype('int64')
response_unemployement= Walmart[Walmart['Store']==1]['Unemployment'].astype('int64')
```

```python
from sklearn.model_selection import train_test_split
x_train_cpi, x_test_cpi, y_train_cpi, y_test_cpi= train_test_split(feature_dataset, response_cpi, random_state=1)
x_train_unemp, x_test_unemp, y_train_unemp, y_test_unemp = train_test_split(feature_dataset,response_unemployement,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg= LogisticRegression(max_iter=10000)
logreg.fit(x_train_cpi, y_train_cpi)
y_pred= logreg.predict(x_test_cpi)
logreg.fit(x_train_unemp, y_train_unemp)
```

```python
y_pred_unemp= logreg.predict(x_test_unemp)
y_pred_unemp
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test_cpi,y_pred))
print(metrics.accuracy_score(y_test_unemp, y_pred_unemp))
```

```python
print('cpi actual:', y_test_cpi.values[0:20])
print('cpi predicted:', y_pred[0:20])
print('actual unemployment:',y_test_unemp.values[0:20])
print('predicted unemployment:', y_pred_unemp[0:20])
```
